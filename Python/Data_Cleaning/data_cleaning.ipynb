```python
#import libraries
import pandas as pd
from datetime import datetime as dt

#load data into dataframe
df = pd.read_csv('store_reviews.csv')

df.head()
```

```python
#drop unwanted columns
df = df.drop(['Profile Link', 'Review Title', 'Review Text'], axis=1)
df.head()
```

```python
#Review Count
#remove " review" and " reviews"
df['Review Count'] = df['Review Count'].str.replace('\D', '', regex=True)

#convert Review Count to int
df['Review Count'] = df['Review Count'].astype(int)

df.head()
```

```python
#Review Date
#split the date and time into two columns
split_rd = df['Review Date'].str.split('T', expand=True)
df["Review Date"] = split_rd[0]
df.insert(4, 'Review Time', split_rd[1])

#remove ".000Z" from Review Time
df['Review Time'] = df['Review Time'].str.replace('.000Z', '', regex=True)

df.head()
```

```python
#Rating
#remove "Rated " and " out of 5 stars"
df['Rating'] = df['Rating'].str.replace('Rated ', '', regex=True)
df['Rating'] = df['Rating'].str.replace(' out of 5 stars', '', regex=True)

#convert Review Count to int
df['Rating'] = df['Rating'].astype(int)

df.head()
```

```python
#Date of Experience
#convert date string to the same format as Review Date
for i in range(df.shape[0]):
    df.loc[i, 'Date of Experience'] = dt.strptime(df.loc[i, 'Date of Experience'], '%d-%b-%y').strftime('%Y-%m-%d')

df.head()
```

```python
#write to csv file
df.to_csv('store_reviews_cleaned.csv', index=False)
```
